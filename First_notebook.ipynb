```python
#Final Challenge 
#For each olympics, which US athlete has won the most total number of medals? Include the discipline
gy = oo[oo.NOC == 'USA']
#want to unstack the medals to sum them later 
gy = gy.groupby(['Edition', 'Athlete', 'Medal']).size().unstack('Medal', fill_value=0)
#create new column in the data frame 
gy['Total'] = gy['Gold'] + gy['Silver'] + gy['Bronze']
gy.reset_index(inplace=True)
tu = [group.sort_values('Total', ascending=False)[:1] for year, group in gy.groupby('Edition')]
tu
#top element in each group
top = pd.DataFrame()
for i in tu:
    top = top.append(i)
top
```
